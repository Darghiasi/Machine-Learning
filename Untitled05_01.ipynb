   - The **green dashed line** represents the **median** popularity score.

   - A **low kurtosis** (~3) means the distribution is similar to a normal distribution.  
   
4. **Key Takeaways:**  
   - If the dataset is highly skewed, **we may need transformations (e.g., log transformation)** before training a regression model.
   - If outliers exist, we may consider **removing or adjusting extreme values** to improve model stability.

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from IPython.display import HTML, display

# --- Assume your data is already loaded and preprocessed ---
# Example:
# train_data = pd.read_csv('train_data.csv')
# test_data = pd.read_csv('test_data.csv')
# target = 'popularity'
# features = ['year', 'bpm', 'nrgy', 'dnce', 'dB', 'live', 'val', 'dur', 'acous', 'spch']
# X_train = train_data[features]
# y_train = train_data[target]
# X_test = test_data[features]

# For demonstration, we'll assume X_train, y_train, and X_test are defined.

# Define the base models
estimators = [
    ('lr', LinearRegression()),
    ('rf', RandomForestRegressor(random_state=42, n_estimators=100)),
    ('gb', GradientBoostingRegressor(random_state=42, n_estimators=100, learning_rate=0.1, max_depth=3)),
    ('xgb', xgb.XGBRegressor(random_state=42, n_estimators=100, learning_rate=0.1, max_depth=3,
                              objective='reg:squarederror'))
]

# Define the final estimator (meta-model)
final_estimator = LinearRegression()

# Create the stacking ensemble
stacking_regressor = StackingRegressor(
    estimators=estimators,
    final_estimator=final_estimator,
    cv=5
)

# Train the stacking ensemble on the full training data
stacking_regressor.fit(X_train, y_train)

# Generate predictions on the test data
ensemble_preds = stacking_regressor.predict(X_test)

# Prepare the submission DataFrame
# Use the 'Id' column if available; otherwise, use the test_data index.
if 'Id' in test_data.columns:
    test_ids = test_data['Id']
else:
    test_ids = test_data.index

submission_ensemble = pd.DataFrame({
    'Id': test_ids,
    'pop': ensemble_preds  # using 'pop' as required
})

# Save the submission CSV file.
submission_filename = 'submission_ensemble.csv'
submission_ensemble.to_csv(submission_filename, index=False)
print(f"Submission file '{submission_filename}' created.")

# Create a download link (for Jupyter Notebook users)
def create_download_link(filename, title="Download file"):
    return HTML(f'<a href="{filename}" download>{title}</a>')

print("\nDownload Link:")
display(create_download_link(submission_filename, "Download submission_ensemble.csv"))
```
