 This notebook provides a structured approach for predicting song popularity using machine learning techniques. Let's begin by loading and inspecting our dataset.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import os
```

 2. Load and Inspect the Data
 Let's read our data using Pandas:

```python
df_train = pd.read_csv("CS98XRegressionTrain.csv")
df_test = pd.read_csv("CS98XRegressionTest.csv")
```

Display dataset shapes

```python
print("Train file shape:", df_train.shape)
print("Test file shape:", df_test.shape)
```

Display first few rows

```python
display(df_train.head())
display(df_test.head())
```

 **Overview of the Data:**
 - The training dataset contains historical song data along with their popularity score (`pop`).
 - The test dataset lacks the `pop` column, which we aim to predict.
 - Features include numerical attributes such as `bpm`, `nrgy`, `dnce`, etc.
 - Some categorical data (`title`, `artist`, `top genre`) are removed as they do not contribute to numerical modeling.

 3. Data Preprocessing
 **Drop irrelevant columns:** `title`, `artist`, `top genre`

```python
columns_to_drop = ['title', 'artist', 'top genre']
df_train = df_train.drop(columns=[col for col in columns_to_drop if col in df_train.columns], errors='ignore')
df_test = df_test.drop(columns=[col for col in columns_to_drop if col in df_test.columns], errors='ignore')
```

**Handle missing values**

```python
df_train.fillna(df_train.median(), inplace=True)
df_test.fillna(df_test.median(), inplace=True)
```

 **Extract target variable**

```python
if 'pop' in df_train.columns:
    y_train = df_train['pop']
    X_train = df_train.drop(columns=['pop'])
else:
    raise ValueError("Error: 'pop' column not found in training data")
```

**Ensure only numerical features are used**

```python
X_train = X_train.select_dtypes(include=[np.number])
X_test = df_test.select_dtypes(include=[np.number])
```

4. Exploratory Data Analysis (EDA)
 **Feature Correlation Heatmap**

```python
plt.figure(figsize=(12, 6))
sns.heatmap(X_train.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()
```

**Distribution of Target Variable**

```python
plt.figure(figsize=(8, 6))
sns.histplot(y_train, bins=30, kde=True, color='blue')
plt.title("Distribution of Song Popularity (pop)")
plt.xlabel("Popularity Score")
plt.ylabel("Frequency")
plt.show()
```

**Feature Distributions and Outliers**

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=X_train, orient='h', palette='coolwarm')
plt.title("Feature Distributions and Outliers")
plt.show()
```

5. Model Training & Evaluation
 **Split the Data for Validation**

```python
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

**Standardize Features**

```python
scaler = StandardScaler()
X_train_split_scaled = scaler.fit_transform(X_train_split)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

**Train and Compare Models**

```python
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42)
}
from sklearn.metrics import root_mean_squared_error
results = {}
for name, model in models.items():
    model.fit(X_train_split_scaled, y_train_split)
    y_pred = model.predict(X_val_scaled)
    rmse = root_mean_squared_error(y_val, y_pred)
    results[name] = rmse
```

**Model Performance Comparison**

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=list(results.keys()), y=list(results.values()))
plt.xlabel("Model")
plt.ylabel("RMSE")
plt.title("Regression Model Performance Comparison")
plt.xticks(rotation=45)
plt.show()
```

6. Model Selection & Final Prediction
 **Selecting the best model**

```python
final_model = RandomForestRegressor(n_estimators=100, random_state=42)
final_model.fit(scaler.fit_transform(X_train), y_train)
```

**Predict on Test Data**

```python
y_test_pred = final_model.predict(X_test_scaled)
df_predictions = pd.DataFrame({"Id": df_test['Id'], "pop": y_test_pred.round().astype(int)})
```

**Save Predictions**

```python
output_file = "spotify_regression_predictions.csv"
if os.path.exists(output_file):
    os.remove(output_file)
df_predictions.to_csv(output_file, index=False)
print(f"Predictions saved to {output_file}")

display(df_predictions.head())
```
